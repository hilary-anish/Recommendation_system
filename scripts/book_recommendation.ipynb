```python
# drop the last two URL columns

book_df = book_df.drop(['Image-URL-M', 'Image-URL-L'], axis=1)
```

```python
rating_df.rename(columns={'User-ID':'User_id','Book-Rating':'Book_rating'}, inplace=True)
rating_df.head(2)
```

```python
users_df.rename(columns={'User-ID':'User_id'}, inplace=True)
users_df.head(2)
```

```python
users_df.isnull().sum()
```

```python
r = rating_df.groupby('User_id')['Book_rating'].count().reset_index().sort_values(by='Book_rating', ascending=False)
r.rename(columns={'Book_rating':'num_of_ratings'}, inplace=True)
```

```python
r
```

```python
user_ids_with_high_rate = r[r['num_of_ratings'] > 200]['User_id'].unique()
final_rating_df = rating_df[rating_df['User_id'].isin(user_ids_with_high_rate)]
final_rating_df
```

- The corresponding User Ids with number of book ratings greater than 200 has been filtered.
- Books with fewer number of ratings cannot contribute in this process

```python
book_df.head(2)
```

```python
book_rate_df = final_rating_df.merge(book_df, on='ISBN')
book_rate_df.head(2)
```

- Inner join is performed between the book_df and rating_df

```python
t = book_rate_df.groupby('Title')['Book_rating'].count().reset_index().sort_values(by='Book_rating', ascending=False)
t.rename(columns={'Book_rating':'num_of_ratings'}, inplace=True)
t
```

```python
book_rate = book_rate_df.merge(t, on='Title')
book_rate.head(2)
```

```python
final_data = book_rate[book_rate['num_of_ratings']>=30]
final_data.shape
```

```python
# Remove duplicates in 'userid' and 'title' column

final_data = final_data.drop_duplicates(subset=['User_id', 'Title'], keep='first')

final_data.shape
```

```python
final_data
```

### Create a Pivot table between User and Rating

```python
collaborative_pivot = final_data.pivot_table(columns='User_id', index='Title', values='Book_rating')
collaborative_pivot.head()
```

```python
collaborative_pivot.shape
```

```python
collaborative_pivot.fillna(0, inplace=True)
```

```python
collaborative_pivot.head()
```

### Sparse and Model Creation

```python
from scipy.sparse import csr_matrix

sparsed_matrix = csr_matrix(collaborative_pivot)
sparsed_matrix
```

```python
from sklearn.neighbors import NearestNeighbors

neighbor_model = NearestNeighbors(algorithm='brute')

neighbor_model.fit(sparsed_matrix)
```

```python
# select the first book with ratings from different user_id

distance, suggestion = neighbor_model.kneighbors(collaborative_pivot.iloc[13,:].values.reshape(1,-1), n_neighbors=5)
```

```python
# minkowski distance metric with respect to different books based on user ratings

distance
```

```python
# corresponding index values of books

suggestion
```

```python
for i in range(len(suggestion)):
    print(collaborative_pivot.index[suggestion[i]])
```

- People who have liked the book "A Bend in the Road" also like "Hurricane Bay","Gates of Paradise", "The Experiment", "Getting Lucky"

```python
# collection of all the books name

book_library = collaborative_pivot.index
```

```python
import numpy as np

def recommend_book(book_name):
    book_id = np.where(book_library == book_name)[0][0]
    distance, suggestion = neighbor_model.kneighbors(collaborative_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=5)
    
    print('The suggested books for further reading are:')
    for i in range(len(suggestion)):
        suggested_books = collaborative_pivot.index[suggestion[i]]
        for j in suggested_books:
            print(j)
        
    
```

```python

recommend_book("Hurricane Bay")
```

```python
def fetch_poster(suggestion):
    url_list = []
    for i in range(len(suggestion)):
        suggested_books = collaborative_pivot.index[suggestion[i]]
        for j in suggested_books:
            print(j)
            img_url = final_data.loc[final_data['Title'] == j, 'Img_url'].values[0]
            url_list.append(img_url)

    print(url_list)


fetch_poster(suggestion)
```
